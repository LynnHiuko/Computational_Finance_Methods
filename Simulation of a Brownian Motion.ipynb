Brownian Motion is the random fluctuation of asset prices over time.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameter for 50 sample paths of a Wiener process
#Total Time
T=1.0
#Total number of steps
n=5
#Number of sample paths
p=50
#Step size
dt=T/n
#To create a time grid
t=np.linspace(0,T,n+1)
```

```python
#Create an array for Brownian paths
#p paths, each of length n+1
W = np.zeros((p, n+1))
```

```python
#Simulating m Brownian paths
#There are n standard normal vars
# Scaling increments to avoid concerns around variance
#Calculate cumulative sum to get the path
for i in range (p):
  Z = np.random.normal(0,1,n)
  increments = np.sqrt(dt) * Z
  W[i,1:] = np.cumsum(increments)
```

```python
#Plot
#W[i] is the i-th Brownian motion path containing n+1 values: W0, Wt,...,WT
plt.figure(figsize = (10,6))
for i in range(p):
  plt.plot(t, W[i], linewidth=1)

plt.title("50 Sample Paths of a Wiener Process")
plt.xlabel("Time t")
plt.ylabel("Wt")
plt.grid(True)
plt.show()
```

```python
#Adjusting n to 500 for accuracy and statistical realism showing independence of increments
#Parameter for 50 sample paths of a Wiener process
T=1.0
#Total number of steps
n=500
p=50
dt=T/n
t=np.linspace(0,T,n+1)

W = np.zeros((p, n+1))

for i in range (p):
  Z = np.random.normal(0,1,n)
  increments = np.sqrt(dt) * Z
  W[i,1:] = np.cumsum(increments)

plt.figure(figsize = (10,6))
for i in range(p):
  plt.plot(t, W[i], linewidth=1)

plt.title("50 Sample Paths of a Wiener Process")
plt.xlabel("Time t")
plt.ylabel("Wt")
plt.grid(True)
plt.show()
```
